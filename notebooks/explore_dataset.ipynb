```python
import nltk

from transformers import BertTokenizer
```

```python
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
```

```python
sample_text = "I have a new GPU!"
```

```python
tokens = tokenizer.tokenize(sample_text)
tokens
```

```python
pos_tags = nltk.pos_tag(tokens=tokens)
pos_tags
```

```python
tokenenized_docs = tokenizer([sample_text], padding="max_length", return_tensors="np")
```

```python
tokenizer.convert_ids_to_tokens(tokenenized_docs["input_ids"][0])
```
