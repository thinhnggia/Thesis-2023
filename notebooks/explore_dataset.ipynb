```python
DATASET_PATH = "src/dataset/ASAP"
DATASET = os.path.join(DATASET_PATH, "training_set_rel3.tsv")
```

```python
training_data = pd.read_csv(DATASET, sep="\t", encoding="ISO-8859-1")
```

```python
from itertools import product

list_traits = ["trait1", "trait2", "trait3", "trait4", "trait5", "trait6"]
list_raters = ["rater1", "rater2"]

combinations = [p for p in product(*[list_raters, list_traits])]
combinations
```

```python
trait_essays = {
    "1": [],
    "2": [],
    "3": [],
    "4": [],
    "5": [],
    "6": [],
    "7": [],
    "8": []
}
for combination in combinations:
    column_name = f"{combination[0]}_{combination[1]}"
    for essay_set in range(8):
        essay_set += 1
        score = training_data.loc[training_data["essay_set"] == essay_set][column_name]
        if score.any():
            trait_essays[str(essay_set)].append(column_name)

trait_essays
```
